```python
import math

# Read the reads on a FastQ file
def readfile(fname):
    readsg = ''
    with open(fname) as fp:
        line = fp.readline()
        while line:
            readsg = readsg + line
            line = fp.readline()
            line = fp.readline()
            line = fp.readline()
            line = fp.readline()
    return readsg

fname = "ecoli.bfast.fastq"
glist = readfile(fname)
glist = glist.split()
print(glist[0:10])

x = glist[0]
y = x.split('_')
print(y)
print(y[2],y[-1])

start = []
end = []
idx = []
for xlist in glist:
    ylist = xlist.split('_')
    start.append(int(math.floor(float(ylist[2])/1024)))
    end.append(int(math.ceil(float(ylist[3])/1024)))
    idx.append(int(ylist[-1], 16))
    
print(start[0:10])
```

```python
from collections import Counter 
from collections import OrderedDict  

right = 0
right_top1 = 0
wrong = 0
wrong_top1 = 0
falsepositive = 0

def most_frequent(List): 
    return max(set(List), key = List.count) 

def arrange_frequent(List):
    result = [item for items, c in Counter(List).most_common() 
                                      for item in [items] * c] 
    return result 

def frequency_elements(List):
    res =  [(el, List.count(el)) for el in List] 
    return list(OrderedDict(res).items()) 

nbatch = 40
for j in range(0,nbatch):
    fname_best = "results_v4/ecoli" + str(j) + ".pos.par"
    print(fname_best)
    res = []
    with open(fname_best) as fp:
        line = fp.readline()
        while line:
            x = line.strip().split(' ')
            x.pop(0)
            x.pop(0)
            res.append(x)
            line = fp.readline()

    x = []
    for i in range(1000):
        x.append([])
    
    for r in res:
        x[int(r[0])].append(int(r[2]))
    
    
    for i in range(1000):
        freq = frequency_elements(arrange_frequent(x[i]))
        idx = i + j*1000
        if len(freq) > 0 and (start[idx] != 0 or end[idx] != 0):
            #print(start[i],end[i],f)
            isright = 0
            isright_top1 = 0
            falsepositive = falsepositive + len(freq) - 1
            for f in freq:
                if (start[idx]-1 <= f[0] and end[idx]+1 >= f[0]) \
                    or (start[idx]+1 >= f[0] and end[idx]-1 <= f[0]):
                        isright = 1
            #elif (start[idx] <= freq[1][0] and end[idx] >= freq[1][0]) \
            #    or (start[idx] >= freq[1][0] and end[idx] <= freq[1][0]):
            #    right = right + 1
            #elif (start[idx] <= freq[2][0] and end[idx] >= freq[2][0]) \
            #    or (start[idx] >= freq[2][0] and end[idx] <= freq[2][0]):
            #    right = right + 1
            if (start[idx]-1 <= freq[0][0] and end[idx]+1 >= freq[0][0]) \
                    or (start[idx]+1 >= freq[0][0] and end[idx]-1 <= freq[0][0]):
                    right_top1 = right_top1 + 1
            else:
                wrong_top1 = wrong_top1 + 1
            
            if isright:
                right = right + 1
            else:
                wrong = wrong + 1
                print(idx,start[idx],end[idx], freq)#freq[2][0])
        #print(start[i],end[i],f)
            #if start[i+j*1000]-100 < f and start[i+j*1000]+100 > f:
            #    right = right + 1
            #elif end[i+j*1000]-100 < f and end[i+j*1000]+100 > f:
            #    right = right + 1
            #else:
            #    wrong = wrong + 1
            #    print(i,start[i],end[i], f)
    
    #print(res[0])
    #print(len(res))
print(right, wrong)
print(float(right)*100/float(right+wrong))
print(float(right_top1)*100/float(right+wrong_top1))
```
