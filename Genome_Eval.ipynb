```python
from IPython.display import clear_output
import math
import numpy as np
nrow = 64
ncol = 64

# Read the reads on a FastQ file
def readfile(fname):
    readsg = ''
    with open(fname) as fp:
        line = fp.readline()
        while line:
            readsg = readsg + line
            line = fp.readline()
            line = fp.readline()
            line = fp.readline()
            line = fp.readline()
    return readsg.split('\n')

fname = 'bp100/bp100'
readsg = readfile('reads/' + fname + ".bfast.fastq")

readsg = readsg
idx = []
position = []
for l in readsg:
    x = l.split('_')
    if x[0] == '@NC':
        position = position + [int(x[2])]
        idx = idx + [int(x[10],16)]
    
print(idx)
print(position)
```
