```python
import numpy as np
# Read the results of one best file
def readpar(fname):
    readsg = ''
    with open(fname) as fp:
        line = fp.readline()
        while line:
            readsg = readsg + line
            line = fp.readline()
    return readsg.split('\n')
```

```python
fname = "human_results_fin/human0_1670000.pos.par"
par_fname = readpar(fname)
```

```python
def getparidx(par_fname,idx,arraynum):
    fileparlist = []
    parlist = [""]*100000    
    maxlen = 0
    
    for par in par_fname:
    #for i in range(10):
    #    par = par_fname[i]
        par_split = par.split(' ')
        #print par_split
        if len(par_split) > 3 and int(par_split[2]) == idx:
            #print(par_split[4], len(parlist))
            #print(parlist[int(par_split[4])])
            parlist[int(par_split[4])] = parlist[int(par_split[4])] + par_split[5] + ' '

    for i in range(10000):
        if len(parlist[i]) > 0:
            parlist[i] = list(np.unique(parlist[i].split()))
            if len(parlist[i]) > maxlen:
                maxlen = len(parlist[i])
            if len(parlist[i]) > 0:
                #print(i,parlist[i])
                fileparlist.append((idx,arraynum,i,parlist[i]))

    return fileparlist,maxlen

getparidx(par_fname,0,0)
```
