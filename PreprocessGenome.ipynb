```python
import math
import numpy as np
nrow = 64
ncol = 64
nrowcol = nrow*ncol

fname = "GCF_000005845.2_ASM584v2_genomic.fna"

fname_out = "ecoli.fna"
f = open(fname_out,'w')


with open(fname) as fp:
    line = fp.readline().strip()
    line = fp.readline().strip()
    print(line)
    while line:
        buf = []
        for i in range(int(nrowcol/4)):
            buf.append(line)
            line = fp.readline().strip()
            if not line:
                break
        buf = ''.join(buf)
        buf = [buf[i:i+nrowcol/4] for i in range(0, len(buf), nrowcol/4)]
        for item in buf:
            f.write(item + '\n')
f.close()
```
