```python
# Read the results of one best file
def readbest(fname):
    readsg = ''
    with open(fname) as fp:
        line = fp.readline()
        while line:
            readsg = readsg + line
            line = fp.readline()
    return readsg.split('\n')
```

```python
fname = "human_results_fin/human0_1670000.pos.best"
best_fname = readbest(fname)
```

```python
def findreadinbest(best_fname,cur_idx = 0):
    cur_array = []
    for line in best_fname:
        match = line.split(',');
        if match[0] != '' and int(match[0]) == cur_idx:
            cur_array.append(line)
    return cur_array
```

```python
from os import listdir
from os.path import isfile, join
path = 'results_human_0/'
list_of_files = [f for f in listdir(path) if isfile(join(path, f))]
#print(list_of_files)

def find_list_of_best_in_read(list_of_files, idx):
    list_of_best = []
    for fname in list_of_files:
        if fname.endswith(".pos.best"):
            best_fname = readbest(path+fname)
            list_of_best.append(findreadinbest(best_fname,idx))
    return list_of_best
```

```python
def find_list_of_best_of_best(list_of_best):
    array_best_count = []
    for cur_best in list_of_best:
        #print(cur_best, array_best_count)
        if cur_best:
            array_best = cur_best[0].split()
            array_best_count.append(len(array_best)-1)
        # Compare with the results
    idx = [i for i, j in enumerate(array_best_count) if j == max(array_best_count)]
    best_of_best = []
    for i in idx:
        best_of_best.append(list_of_best[i])
    
    if max(array_best_count) - 1 > 0:
        idx = [i for i, j in enumerate(array_best_count) if j == max(array_best_count)-1]
        for i in idx:
            best_of_best.append(list_of_best[i])

    return best_of_best
```
