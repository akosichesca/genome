```python
fname = "bp100_sub0.1.pos_v3b"

nread = []
nshift = []
narray = []
ncolumn = []
nedit = []
tag = [False] * 10
idxs = 10

print(tag)

```

```python
with open(fname) as fp:
    line = fp.readline()
    cnt = 1
    while line:
        l = line.strip().split()[1:]
        tag[int(l[0])-idxs] = True
        line = fp.readline()

print(tag)
```

```python
fname = "bp100_sub0.1.pos_v3b.par"

nread = []
nshift = []
narray = []
ncolumn = []
nedit = []

with open(fname) as fp:
    line = fp.readline()
    cnt = 1
    while line:
        l = line.strip().split()[2:]
        if tag[int(l[0])-idxs] == False: 
            nread.append(int(l[0]))
            nshift.append(int(l[1]))
            narray.append(int(l[2]))
            ncolumn.append(int(l[3]))
            nedit.append(int(l[4]))
        line = fp.readline()
        cnt += 1
```
